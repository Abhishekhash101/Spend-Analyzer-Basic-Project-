```python
import pandas as pd
```

```python
dataframe=pd.read_csv('transactions.csv')
dataframe
```

```python
for i in pd.unique(dataframe['sender']):
    print(f"{i}");
```

```python
dataframe=dataframe[dataframe['merchant']!='UNKNOWN']
dataframe=dataframe[dataframe['txn_type']!='unknown']
```

```python
dataframe[dataframe['merchant']=='MR']['raw_message']
```

```python
dataframe['merchant']=dataframe['merchant'].apply(lambda x: x.split('UPI')[0])
```

```python
def cleaning_merchant(x: str):
    # If 'LINKEDTOVPA' is present, take the part after it
    parts = x.split('LINKEDTOVPA')
    if len(parts) > 1:
        return parts[1].strip()
    else:
        return x.strip()
```

```python
dataframe['merchant'] = dataframe['merchant'].apply(cleaning_merchant)
```

```python
pd.unique(dataframe['merchant'])
```

```python
def get_bank_name(x: str):
    parts = x.split('-')
    if len(parts) > 1:
        return parts[1].strip()
    else:
        return None
```

```python
dataframe['Bank']=dataframe['sender'].apply(get_bank_name)
```

```python
dataframe['Bank'].unique()
```

```python
dataframe=dataframe[dataframe['Bank']!=None]
```

```python
dataframe.to_csv('transactions_cleaned.csv')
```
